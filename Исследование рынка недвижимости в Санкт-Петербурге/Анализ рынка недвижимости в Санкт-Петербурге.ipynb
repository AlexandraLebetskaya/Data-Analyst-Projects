## Шаг 3. Внесение недостающих данных

### Цена квадратного метра

```python
df['price_m'] = df['last_price'] / df['total_area']
df['price_m'] = df['price_m'].astype('int')
```

### День недели, месяц и год публикации объявления

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

```python
def weekday(row):
    weekday = row
    if weekday == 0:
        return 'воскресенье'
    if weekday == 1:
        return 'понедельник'
    if weekday == 2:
        return 'вторник'
    if weekday == 3:
        return 'среда'
    if weekday == 4:
        return 'четверг'
    if weekday == 5:
        return 'пятница'
    return 'суббота'
df['weekday_new'] = df['weekday'].apply(weekday)
```

### Этаж квартиры; варианты — первый, последний, другой

```python
def floor_room(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'послений'
    return 'другой'
df['floor_room'] = df.apply(floor_room, axis = 1)
df.head(3)
```

### Соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
df['livin_total_area'] = df['living_area'] / df['total_area']
df['kihchen_total_area'] = df['kitchen_area'] / df['total_area']
df.head(3)
```

```python
def pivot (data, index, x_name, y_name, name):
    pivot = data.pivot_table(index = index, values = 'price_m', aggfunc = ['count', 'mean', 'median']).reset_index()
    pivot.columns = [index, 'count', 'mean', 'median']
    display('Коэффициент корреляции', data[index].corr(data['price_m']))
    display(pivot)
    
    pivot.plot(y = 'median', x = index)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.title(name)
    plt.show() 
```

```python
pivot(df_good, 'total_area', 'Общая площадь', 'Стоимость квадратного метра', 'Зависимость стоимости квадратного метра от площади')
```
